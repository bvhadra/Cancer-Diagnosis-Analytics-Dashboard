## **Hypotheses**
- Older patients may have a higher mutation count.
- Higher mutation counts may be linked to lower survival rates.
- Certain cancer types may respond better to treatments.

Let's begin by loading and exploring the dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Set visualization style
sns.set(style="whitegrid")

# Load dataset
file_path = "METABRIC_RNA_Mutation.csv"
df = pd.read_csv(file_path)

# Display basic dataset info
df.info()
df.head()
```

## **Summary Statistics**
We generate descriptive statistics for numerical variables to understand their distribution.

```python
# Summary statistics of numerical columns
df.describe()
```

## **Feature Distributions**
We visualize key numerical variables to understand their spread.

```python
# Histograms for key numerical features
num_cols = ["age_at_diagnosis", "mutation_count", "overall_survival_months"]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(num_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color="blue")
    axes[i].set_title(f"Histogram of {col}")

plt.tight_layout()
plt.show()

# Interactive histograms using Plotly
for col in num_cols:
    fig = px.histogram(df, x=col, nbins=30, title=f"Distribution of {col}", marginal="box")
    fig.show()
```
