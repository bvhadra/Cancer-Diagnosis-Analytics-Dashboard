# **Statistical Analysis**
## **Objective**
This notebook performs statistical analysis to identify key trends, correlations, and validate hypotheses.

## **Key Steps**
- Descriptive statistics
- Hypothesis testing
- Optimized correlation analysis

Let's start by loading the dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy.stats import ttest_ind, chi2_contingency

# Load dataset
file_path = "METABRIC_RNA_Mutation.csv"
df = pd.read_csv(file_path)

# Display initial dataset shape
df.shape
```

## **Descriptive Statistics**
We summarize the dataset with key statistical measures.

```python
# Summary statistics of numerical variables
df.describe()
```

## **Optimized Correlation Analysis**
To improve performance, we select a subset of key numerical features for correlation analysis.

```python
# Select key numerical features for correlation
selected_features = ["age_at_diagnosis", "mutation_count", "overall_survival_months", "lymph_nodes_examined_positive", "nottingham_prognostic_index"]
corr_matrix = df[selected_features].corr()

# Plot optimized correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Optimized Correlation Heatmap")
plt.show()
```

## **Hypothesis Testing**
We perform statistical tests to validate assumptions.

```python
# T-test: Compare mutation count between chemotherapy and non-chemotherapy patients
chemo_yes = df[df["chemotherapy"] == 1]["mutation_count"]
chemo_no = df[df["chemotherapy"] == 0]["mutation_count"]
t_stat, p_value = ttest_ind(chemo_yes, chemo_no, nan_policy='omit')
print(f"T-test for Mutation Count & Chemotherapy: t-stat = {t_stat:.2f}, p-value = {p_value:.4f}")

# Chi-square test: Relationship between cancer type and hormone therapy
contingency_table = pd.crosstab(df["cancer_type"], df["hormone_therapy"])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square test for Cancer Type & Hormone Therapy: chi2 = {chi2:.2f}, p-value = {p:.4f}")
```

## **Summary & Conclusions**
- Mutation count varies significantly based on chemotherapy.
- Cancer type and hormone therapy appear to be dependent.
- The optimized correlation analysis highlights key relationships efficiently.

This statistical analysis provides valuable insights for further modeling.
